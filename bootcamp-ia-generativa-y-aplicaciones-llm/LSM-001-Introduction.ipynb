## â“ Frequently Asked Questions

### Can I use LangSmith without LangChain?
**Absolutely!** LangSmith is framework-agnostic and works with any LLM application, whether you're using LangChain, direct API calls, or other frameworks.

### Is my data secure with LangSmith?
**Yes.** LangSmith encrypts all data in transit and at rest. You control what data is sent to LangSmith, and you can configure sampling rates to limit data collection.

### Does LangSmith slow down my application?
**No.** LangSmith is designed for production use with minimal performance overhead. Tracing happens asynchronously and doesn't block your application.

### Can I self-host LangSmith?
**Yes.** LangSmith offers self-hosting options for enterprise customers who need to keep data within their infrastructure.

### How much does LangSmith cost?
LangSmith offers a generous free tier perfect for learning and small projects. Paid plans scale with your usage and team size.
