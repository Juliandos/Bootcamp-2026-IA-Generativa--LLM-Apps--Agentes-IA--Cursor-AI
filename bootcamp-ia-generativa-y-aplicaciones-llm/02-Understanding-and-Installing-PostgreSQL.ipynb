## Basic SQL Commands ðŸ“

SQL (Structured Query Language) is how we communicate with PostgreSQL. Here are the essential commands:

### 1. Creating a Table

Let's create a simple test table to practice:

````python
# Create a test table
!PGPASSWORD=pass12 psql -U user12 -h localhost -d mydatabase -c "
CREATE TABLE test_todos (
    id SERIAL PRIMARY KEY,
    name TEXT NOT NULL,
    completed BOOLEAN DEFAULT FALSE
);"
````

**What each part means:**
- `CREATE TABLE test_todos`: Create a new table named "test_todos"
- `id SERIAL PRIMARY KEY`: Auto-incrementing unique identifier
- `name TEXT NOT NULL`: Text field that can't be empty
- `completed BOOLEAN DEFAULT FALSE`: True/false field, defaults to false

### 2. Inserting Data (CREATE)

````python
# Add some test data
!PGPASSWORD=pass12 psql -U user12 -h localhost -d mydatabase -c "
INSERT INTO test_todos (name, completed) VALUES 
('Buy groceries', false),
('Walk the dog', true),
('Learn PostgreSQL', false);"
````

### 3. Reading Data (READ)

````python
# Show all data in our test table
!PGPASSWORD=pass12 psql -U user12 -h localhost -d mydatabase -c "SELECT * FROM test_todos;"
````

````python
# Show only incomplete todos
!PGPASSWORD=pass12 psql -U user12 -h localhost -d mydatabase -c "SELECT * FROM test_todos WHERE completed = false;"
````

### 4. Updating Data (UPDATE)

````python
# Mark "Learn PostgreSQL" as completed
!PGPASSWORD=pass12 psql -U user12 -h localhost -d mydatabase -c "UPDATE test_todos SET completed = true WHERE name = 'Learn PostgreSQL';"
````

````python
# Check that it was updated
!PGPASSWORD=pass12 psql -U user12 -h localhost -d mydatabase -c "SELECT * FROM test_todos;"
````

### 5. Deleting Data (DELETE)

````python
# Delete a specific todo
!PGPASSWORD=pass12 psql -U user12 -h localhost -d mydatabase -c "DELETE FROM test_todos WHERE name = 'Walk the dog';"
````

````python
# Check what's left
!PGPASSWORD=pass12 psql -U user12 -h localhost -d mydatabase -c "SELECT * FROM test_todos;"
````

## ðŸ“Š SQL Summary (CRUD Operations)

| Operation | SQL Command | Example | What it does |

Instead of hardcoding database credentials in our code (which would be insecure), we use environment variables. Let's check what our app expects:

### ðŸ¤” Why Use Environment Variables?

**âŒ Bad Practice:**
```python
# Never do this - passwords in code!

```

**âœ… Good Practice:**
```python

```

**Benefits:**
- ðŸ”’ **Security**: No passwords in source code
- ðŸ”„ **Flexibility**: Different settings for development/production
- ðŸ“ **Version Control Safe**: .env files are ignored by git

# ðŸ”§ Troubleshooting Common Issues

Having problems? Don't worry! Here are solutions to the most common PostgreSQL issues beginners encounter.

## ðŸš¨ Installation Issues

### Problem: "psql: command not found"

**Cause**: PostgreSQL is not installed or not in your system PATH

**Solutions by Platform:**

#### macOS:
```bash
# If installed with Homebrew, add to PATH
echo 'export PATH="/usr/local/pgsql/bin:$PATH"' >> ~/.zshrc
source ~/.zshrc

# Or try the full path
/usr/local/pgsql/bin/psql --version
```

#### Windows:
1. **Find PostgreSQL installation** (usually `C:\Program Files\PostgreSQL\15\bin`)
2. **Add to PATH**:
   - Open System Properties â†’ Advanced â†’ Environment Variables
   - Edit "Path" system variable
   - Add: `C:\Program Files\PostgreSQL\15\bin`
3. **Restart** Command Prompt

#### Linux:
```bash
# Install PostgreSQL client tools
sudo apt install postgresql-client

# Or for RedHat/CentOS
sudo dnf install postgresql
```

## ðŸ”Œ Connection Issues

### Problem: "connection refused" or "server doesn't exist"

**Cause**: PostgreSQL service is not running

**Solutions:**

#### macOS (Homebrew):
```bash

brew services list | grep postgresql

# Start service
```

#### Windows:

2. **Find** "postgresql-x64-15" service
3. **Right-click** â†’ Start
4. **Set to Automatic** if you want it to start on boot

#### Linux:

```bash
# Enable on boot

```

## ðŸ” Authentication Issues

### Problem: "password authentication failed"

**Cause**: Wrong username/password or user doesn't exist

**Solutions:**

#### Reset postgres superuser password:
```bash
# macOS (Homebrew)
psql postgres -c "ALTER USER postgres PASSWORD 'newpassword';"

# Linux
sudo -u postgres psql -c "ALTER USER postgres PASSWORD 'newpassword';"

# Windows (as Administrator)
psql -U postgres -c "ALTER USER postgres PASSWORD 'newpassword';"
```

#### Recreate our application user:
```bash
# Connect as superuser first
psql -U postgres

# Then run these commands:
DROP USER IF EXISTS user12;
CREATE USER user12 WITH PASSWORD 'pass12';
GRANT ALL PRIVILEGES ON DATABASE mydatabase TO user12;
GRANT ALL ON SCHEMA public TO user12;
```

### Problem: "role does not exist"

**Cause**: User hasn't been created yet

**Solution**: Create the user following our setup steps above

## ðŸ—„ï¸ Database Issues

### Problem: "database does not exist"

**Solutions:**

#### Check existing databases:
```bash

psql -U postgres -l
```

#### Create our database:
```bash
# Create the database
```

### Problem: "permission denied for schema public"

**Cause**: PostgreSQL 15+ changed default permissions

**Solution:**
```bash
psql -U postgres -d mydatabase -c "GRANT ALL ON SCHEMA public TO user12;"
```
