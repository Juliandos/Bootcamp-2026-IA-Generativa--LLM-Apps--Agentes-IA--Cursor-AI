# Gen AI Disruption of SaaS Businesses and Applications

## Two Important Warnings
* On September 2024, Klarna announces it has internally replaced SaaS providers like Salesforce and Workday by Gen AI solutions.
* On November 2024, the CEO of Choco (one of the most popular European Unicorn startups), declares that AI Agents will kill SaaS Applications.

## Klarna replaces SaaS providers with internally-built Gen AI solutions
* **As soon as ChatGPT was launched, the CEO of Klarna (Sebastian Siemiatkowski) contacted with Sam Altman and told him that Klarna, a small Swedish bank, wanted to be their number one ginea pig**.
* From that moment on, Klarna started to experiment with ChatGPT in all their departments to find areas where it could provide value.
* **One of the first areas they saw a huge value was Customer Support**. In February of 2024, Klarna presented its new AI Agent for customer service, stating that it performed the work of 700 customer service agents and handled 2.3 million interactions in its first month of operation. The AI Agent was made in collaboration with OpenAI. Klarna was one of the first clients for OpenAI’s enterprise ChatGPT package, and claims that 90 percent of its workforce consult the tool every day to automate various processes on the job. 
* **On September 2024, Klarna announces it has internally replaced SaaS providers like Salesforce and Workday by Gen AI solutions**.
* The fintech firm Klarna is severing its relationships with two of the biggest enterprise software providers in favor of automating its services with AI. And the company says it could potentially eliminate more.
* Klarna is no longer using Salesforce, a platform that aggregates and packages sales and marketing data for businesses. The company has also removed the HR and hiring platform Workday from its tech stack.
* “We have a number of large internal initiatives that combine AI, standardization, and simplification to enable us to shut down several software-as-a-service providers,” said a Klarna spokesperson.
* Slashing Workday and Salesforce is part of a broader purge of third-party SaaS providers. Klarna intends to replace the programs with its own, internally built applications, ostensibly crafted on OpenAI’s infrastructure.
* **Klarna reduced its workforce by 1,200 workers over the past year. Klarna CEO, Sebastian Siemiatkowski, told the Financial Times on August 2024 that the firm could benefit by reducing its headcount from 3,800 to 2,000 employees**. He has insisted downsizing wouldn’t slow growth as the company leans into AI.
#### Explanation
* Klarna CEO explained in a recent Venture Capital podcast that one of the reasons they did this was because **simplifying the stack of software tools they use makes it easier for their Gen AI stack to work**. He also explained that this approach is a competitive advantage of the company and that they do not want to share all details about it.
* He also shared a second interesting reason: **since Salesforce and Workday have their APIs closed behind a login, we were not able to ask chatGPT to study and learn from these APIs**. But if you use open source alternatives or companies like Stripe that have their API publicly available, chatGPT can access their APIs freely and you can also contribute with code to include features that are relevant for you.
* At the end of the day, Salesforce and Workday stored Klarna Sales and HR data in their databases and then performed different operations with it.
* Now, our guess is that Klarna is using some non-disclosed open source alternatives plus their own Gen AI. To some extent, it seems that they are using Gen AI to talk directly to their databases and to perform the same operations.
* **Notes to consider**:
    * Gen AI Apps are not as accurate as traditional SaaS apps, but they can be improved with user feedback.
    * It is probable that the Gen AI Apps developed internally by the Klarna team are not as advanced as Salesforce and Workday... yet.
    * Another thing to consider is that in order to replace SaaS tools like Salesforce and Workday you not only need to be able to replace the functionality with Gen AI, but also maintaining it. This would take resources and focus from you Engineering team that you may not want to lose from your core product.
#### Implications
* In the near future, many closed-API SaaS apps could be replaced by open-API Saas apps + Gen AI solutions.
* This is a huge threat to closed SaaS companies.
* **This is a huge opportunity for Gen AI Engineers** able to build the alternative solutions to those SaaS apps.
* This will also mean a huge change in the job market. Gen AI Engineers will be the most wanted professionals.

## The CEO of one of the most popular Unicorns declares that "AI Agents will kill SaaS Apps" and tells how they are seeing this change in their company.
* **Daniel Khachab is the co-founder and CEO of Choco**. Today, Choco’s AI platform facilitates half of all food traded in major cities like New York, Paris, London, and Berlin, cutting food waste and streamlining distribution. Since its founding in 2018, Choco has raised $330 million from Bessemer, Coatue (its first European investment), and Insight, reaching unicorn status within 2.5 years.
* **Previously, Daniel was one of the Managing Directors at Rocket Internet**, a one of the top startup incubators worldwide.
* In November 2024, Khachab was interviewed in a popular Venture Capital podcast.
* "**I think (traditional) SaaS is dead**. We are going to a new era that we can call the agentification of SaaS". Choco went from a SaaS model to their actual model where they claim 100% of their income is coming from AI.
* "When ChatGPT was launched I realized it was going to learn how to code very soon. And by then, when is our technological moat is going to be? Using Gen AI, anyone could replicate in a question of days what we have built in many years."
* "**3 things made us go AI-first:**
    * The threat of someone using Gen AI to replicate our SaaS App and the realization that we ourselves could use Gen AI to radically improve our model and disrupt our competitors.
    * The realization that company building was going to change fundamentally because so much value is going to come out of the API, and therefore there were so much new skills to learn for every single role in our company.
    * The realization that we may not know how the future is going to be like, but the best way to be prepared for it is to learn how to build with Gen AI on every single part of the organization".
* **The transition from SaaS to AI**:
    * **UI**: in a traditional SaaS app, you use menus, links, buttons, forms, emails, downloads, etc, to make the app do things for you. In an AI app, you just use natural language and the UI designer will define the personality of that AI Agent: serious, funny, casual, flirting, etc.
        * The main interface will be conversational, multimodal chat.
        * And that interface will have to include a way for the user to provide feedback that improve the quality of the AI Agent responses.
    * **Accuracy**: traditional SaaS is 100% accurate, Generative AI Apps are not. The user will have to help the AI App to be better with his feedback. The user will have to educate the AI App in the same way you educate a new junior employee.
    * **Marketing and Sales pitch to the customer**: in a traditional SaaS model, you ask the customer what problem does he have and then you build a solution for it. In an AI model, you say the customer "here you have this AI Agent able to perform all these skills, it can be like an employee for you, but he will not take vacation, time off, sick days, etc, and you can hire as many of them as you need at any moment according to your fluctuating needs..."
    * **Engineering**: from traditional software development to AI-assisted software development, from traditional QA to AI-assisted QA, etc.
    * **Customer support**: from just humans to AI-assistants or AI-Assistants + Humans for escalated questions.
    * **HR, Finance, etc**: from traditional roles to AI-assisted/replaced roles. All these administrative tasks are going to change radically: for example, you do not need to make calculations in a spreadsheet, just ask for the result to the AI Agent.
    * **Data entry**: from keyboard to voice, images, and handwritten notes and drawings (multimodality). For example, instead of creating and updating sales leads in a traditional way, the sales rep may only send a voice message to the AI Agent after the sales call and the AI Agent will update the lead record.
