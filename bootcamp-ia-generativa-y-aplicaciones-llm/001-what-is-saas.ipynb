# Why SaaS?

## What is a SaaS Business?
Many things explain SaaS:
* The SaaS Business Model.
* The SaaS Product.
* The SaaS Pricing.
* The SaaS Metrics.
* The SaaS Team: Developers, Content Marketing Lead Generation Funnel, and Sales (SDRs and Sales Reps).

## Origins of SaaS
* Old software products:
    * Software in a disk.
    * Buy a license forever or for one year. Expensive. Rigid.
    * Manual updates.
* New Software As a Service (SaaS) products:
    * 1999: Salesforce. 
    * Software in the cloud. Accesible from any device with internet connection. Automatic updates, agile development. Good data security.
    * Monthly subscriptions. Cheap. Flexible.
    * SaaS startups.

## Cloud services providers make SaaS possible
Some of the most popular:
* AWS.
* Google cloud.
* Microsoft Azure.

## Today, SaaS businesses are everywhere
* Subscription-based apps in your computer, cell phone, smart watch, etc.

# The SaaS challenges

## SaaS startups follow a growth cycle different from most SMBs
* Customer Development: confirm the customer problem.
* MVP: confirm your product solves the customer problem.
* Product-Market fit: confirm the customer is willing to pay your price.
* Venture Capital: investment to grow fast via Lead Gen and Sales.

## Main tech challenges
* Integration with 3rd party APIs.
* Data security.
* Scalability.

## Main business challenges
* To think about the startup hype and not about the customer.
* Timing: entering in the market too soon or too late.
* Market too broad.
    * Solution: start focusing on a small niche and do not scale until you have optimized your performance there.
* Product-Market fit.
    * Solution: direct contact with the market.
* New product.
    * Solution: market evangelization. Takes time. 
* New company. No success cases.
    * Solution: special promotions for early users.
* Small initial revenue vs. Startup costs (developers, marketing, sales, etc.).
    * Solution: seed capital... or bootstraping.
* Scalability.
    * Scale development team and process.
    * Scale lead generation marketing and sales teams.
    * Trying to scale when giant competitors have already entered in your market. If Google enters in your industry, you will be soon dead or sold.
* Metrics performance.
    * Solution: analytics system.  

# Main steps, components, and keys of success

## Go where the customer is
* Traditional product development:
    * assume a problem.
    * write a biz plan.
    * persuade potential customers.
    * build product.
    * receive feedback... too late.
* Lean product development: 
    * go where the customer is and find a big problem that your (new) technology can solve. Test your hypotheses, confirm that the customer really feels this pain.
    * what value is the customer looking for?
    * what are the essential components to get this value? focus. validate your ideas with the customer before starting to develop. Less features, olympic quality.
    * stay in close contact with customers to develop the right product. Build the MVP in several iterations. The customer feedback is the main factor in building the MVP.
    * frugality, flexibility, speed.
    * no biz plan, perhaps a busines model canvas.
        * niche of customers.
        * core problems.
        * core product features.
        * value proposition.
        * unique advantage.
        * channels to customers.
        * key metrics to track.
        * core cost structure.
        * core revenue streams.
    * continuous improvement.

## Master the art of war
* Innovation gives you monopolistic power, at least until your competitors can imitate you.
* Look for business models with moats to prevent competition.
* Only the paranoids survive. Keep your secrets secret.
* Look for growing markets and long term trends.
* Analyze your position vs. direct and indirect competitors: strengths, weaknesses, opportunities, threats (SWOT analysis).
* Do not think too hard, the market will tell you soon if you need to pivot.

## Lauch the bait and see if there are fishes in the pond
* Communicate your idea to customers in the niche: blog, interviews, email, etc.
* See if they are interested in knowing more or in trying the beta version: get their contact info and use them as customer-teachers.

## Lean Market Research
* The key is the direct and frequent contact with the customer.
* Macro market analyis comes last, not first.

## Avoid distractions
* In the initial stage, distractions are everything other than:
    * Talking to customers.
    * Building product.
* Distraction sources:
    * Business Incubator.
    * Co-founders.
    * Investors.
    * Wrong early adopters.
    * Events.
    * PR.
    * The next investment round.
    * The next big thing.

## The dangers of co-founders, investors, and team
* Do not add a team member until you have a more-than-clear necessity.
* Do not look for investment you have a more-than-clear necessity.
* Do not assume that anyone knows about one particular role until he has demonstrated it to you.
* Do not add a co-founder for moral support or to impress investors.

## Agile methodology
* Core idea: quick and iterative product development.
* Variations: scrum, kanban, extreme programming...
* Common patterns:
    * Daily stand-ups
    * Development sprints to build feature, improvement, etc

## MVP
* Quick and cheap.
* Draft (from landing page to demo video to toy demo).
* Solves core problem of customer niche.
* The customer is interested (from "I want to know more" to "I am willing to pay for this").

## Beta version
* Test with customers. Get feedback.
* Refine product.
* Test again.

## Product Value Proposition
* what problem does it solve.
* what value does it provide.
* why is it better than the alternatives.

## Product-Market Fit
* Customer problem confirmed.
* Product solves problem. Confirmed.
* Customer is willing to pay for the price of the product. Confirmed.
* Customer is willing to stay with the product. Confirmed.

## When have you reached the product quality you need?
* When you customer feels the necessity to call a friend and recommend him your product.
* When your customer does not "like" your product, but "LOVES" your product.

## SaaS Pricing
* Subscription-based.
* Monthly and Year plans.
* 3 to 4 tiers.
* Usually, some freemium strategy.
    * Basic plan free.
    * Or free-trial during a period of time.

## SaaS Metrics
* Lead Conversion Rate.
* Customer Acquisition Cost.
* Monthly Recurring Revenue.
* Customer Lifetime Value.
* Churn Rate.
* Customer Retention Cost.

## Do not abandon your technical advantage
* Your value is to master the innovative technology. As long as you do not abandon this (and if you are CEO you will be tempted to abandon it for other biz tasks), you will remain in a very good position for the future.
