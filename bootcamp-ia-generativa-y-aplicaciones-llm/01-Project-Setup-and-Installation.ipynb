### ğŸ”§ Troubleshooting Common Issues

**Problem**: Frontend shows "Connection Error" or todos don't save
- **Solution**: Make sure the backend server is running on port 8000
- **Check**: Visit `http://localhost:8000/docs` to confirm the API is working

**Problem**: "Module not found" errors
- **Solution**: Make sure you ran `npm ci` (not `npm install`) in the frontend directory
- **Solution**: Make sure you ran `pip install -r requirements.txt` in the backend directory

**Problem**: Database connection errors
- **Solution**: Make sure PostgreSQL is running: `brew services start postgresql`
- **Solution**: Check that the database exists: `psql -l | grep mydatabase`

**Problem**: Port already in use
- **Backend**: Change port with `uvicorn main:app --reload --port 8001`
- **Frontend**: Next.js will automatically suggest a different port

## ğŸ“‚ Backend Structure (`001-fastapi-backend/`)

Key files and their purposes:

- **`main.py`**: Main application file, defines API endpoints
- **`config.py`**: Configuration settings (database connection, etc.)
- **`database.py`**: Database connection setup
- **`models.py`**: Database table definitions (Todo model)
- **`schemas.py`**: Data validation schemas (what data looks like)
- **`crud.py`**: Database operations (Create, Read, Update, Delete)
- **`routers/todos.py`**: Todo-specific API routes
- **`requirements.txt`**: Python package dependencies
- **`.env`**: Environment variables (database credentials)
- **`alembic/`**: Database migration files

## ğŸŒ Frontend Structure (`002-nextjs-frontend/todo-app/`)

Key files and their purposes:

- **`pages/index.js`**: Main page of the web app
- **`components/layout.js`**: Overall page layout and styling
- **`components/todo-list.js`**: Main todo list component with API integration
- **`components/todo.js`**: Individual todo item component
- **`styles/`**: CSS styling files
- **`package.json`**: Node.js package dependencies and scripts
- **`.env`**: Environment variables (API URL)

## ğŸ”„ How It All Works Together

1. **User interacts** with the React frontend (add/complete/delete todos)
2. **Frontend sends HTTP requests** to the FastAPI backend
3. **Backend processes the request** and updates the PostgreSQL database
4. **Backend sends response** back to the frontend
5. **Frontend updates the UI** to show the changes
