# What is Claude Code and why today it is the top contender for Cursor?

## What is Claude Code?
* [Website of Claude Code](https://www.anthropic.com/claude-code).
* Cursor Alternative built by the creators of Claude, the top coding LLM.
* Hugely popular after the launch of Claude 4.
* It uses a different approach from Cursor and similar alternatives (Windsurf, etc).
* It requires a monthly subscription. [See Pricing Plans](https://www.anthropic.com/claude-code).

## Why it has become the top alternative to Cursor today?
* Cursor and similar alternatives initially were clever wrappers around the Claude LLM model.
* As we were anticipating, Anthropic (the company behind the Claude LLM model) has decided to go after the multi-million dollar market Cursor is serving.

## What are our impressions about Claude Code?
* It is a very powerful option.
* It has the ingredients to become the dominant Coding Assistant, but we think it probably will evolve to be more user-friendly. It is still in a very early stage.
* We personally do not like to use it in the terminal by itself, out of a code editor.
* We prefer to use it in combination with Cursor. This is an expensive option, but if it makes (cost/value) sense for you it is the best alternative today in our opinion. We will tell you why we choose this combination and how we use it.
